```python
from pandas.io.json import json_normalize
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
config = json.loads(open('combinedArray.json').read())
```

```python
favLang = json_normalize(config)
```

```python
favLang = favLang[['city', 'country', 'favLanguage', 'gender', 'nativeLanguage', 'programmingExperience.years', 'programmingExperience.months']]
```

```python
favLang
```

```python
favLang = favLang.replace(["United States", "United States of America"], "USA")
favLang = favLang.replace("Brasil", "Brazil")
```

```python
favLang['country'].value_counts()
```

```python
favLang = favLang.replace("c++", "C++")
favLang = favLang.replace(["Javascript", "ECMAScript 6", "Javascript - NodeJS", "javascript", "JS"], "JavaScript")
favLang = favLang.replace([]"java", "Java")

favLang['favLanguage'].value_counts()
```
